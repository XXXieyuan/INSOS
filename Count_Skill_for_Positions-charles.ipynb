```python
def convert_time(rdate):
    numdict = {
        "zero": '0',
        "one": "1",
        "two": '2',
        "three": '3',
        'four': '4',
        'five': '5',
        'six': '6',
        'seven': '7',
        'eight': '8',
        'nine': '9',
        'ten': '10',
        'eleven': '11',
        'twelve': '12',
        'thirteen': '13',
        'fourteen': '14',
        'fifteen': '15',
        'sixteen': '16',
        'seventeen': '17',
        'eighteen': '18',
        'nineteen': '19',
        'twenty': '2',
        'thirty': '3',
        'forty': '4',
        'fifty': '5',
        'sixty': '6',
        'seventy': '7',
        'eighty': '8',
        'ninety': '9',

    }
    check = " second or seconds or minute or minutes or hours or hour or day or days or month or months or ago"
    tis = ['twenty', 'thirty', 'forty', 'fifty',
           'sixty', 'seventy', 'eighty', 'ninety']
    memory = []
    for i in range(0, len(rdate)):

        if rdate[i] in check:
            if rdate[i] == 'day':
                rdate[i] = 'days'
                break
            elif rdate[i] == 'hour':
                rdate[i] = 'hours'
                break
            elif rdate[i] == 'month':
                rdate[i] == 'months'
                break
            elif rdate[i] == 'minute':
                rdate[i] == 'minutes'
                break
            elif rdate[i] == 'ago':
                break

        else:

            memory.append(rdate[i])
            rdate[i] = numdict[rdate[i]]

    rdate = ' '.join(rdate)

    # delete space

    for i in range(0, len(rdate)):
        flag = False
        if rdate[i] == ' ':
            if rdate[i-1].isdigit() and rdate[i+1].isdigit():
                rdate = rdate[:i] + rdate[i+1:]
                break
            elif rdate[i-1].isdigit() and rdate[i+1].isdigit() == False:
                for mem in memory:
                    if mem not in tis and len(memory) == 1:
                        flag = True
                if flag == False:

                    rdate = rdate[:i] + '0' + rdate[i:]
                    break

    return rdate
```

```python
import qrcode
from PIL import Image
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=10,
    border=4,
)
qr.add_data('https://international-student-it-jobs.azurewebsites.net')
qr.make(fit=True)
img = qr.make_image(fill_color="black", back_color="white").convert('RGB')
logo_display = Image.open('test.png')
logo_display.thumbnail((60, 60))
logo_pos = ((img.size[0] - logo_display.size[0]) // 2, (img.size[1] - logo_display.size[1]) // 2)
img.paste(logo_display, logo_pos)
img.save("sample2.png")
```
